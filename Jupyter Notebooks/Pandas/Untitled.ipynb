```python
import pandas as pd
import requests
```

```python
# getting TV series from TMDB and converting to dfb

df = pd.DataFrame()
API_KEY = "YOUR_API_KEY"
for i in range(142):
    temp = requests.get(f"https://api.themoviedb.org/3/tv/top_rated?api_key={API_KEY}&language=en-US&page={i+1}").json()['results']
    temp = pd.DataFrame(temp)
    df = pd.concat([df,temp], ignore_index=True)
```
