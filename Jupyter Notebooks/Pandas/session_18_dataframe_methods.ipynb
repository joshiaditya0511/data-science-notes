## rank() function

```python
# rank() takes a column or row in a df or series and then ranks the indices according to the values in the given column/row.

# For example, if you pass a column containing numeric values, it will assign rank 1 to the lowest value and -
# - assign lowest rank to the highest value. (You can obviously change that behaviour by changing bool value of -
# the 'ascending' parameter)

s = pd.Series(range(2,22,2), index=list("abcdefghij"))
s["g"] = s["b"]
s['i'] = s['b']
s
```

```python
# The 'method' parameter is used to specify how to rank conflicting values / equal values.

# The 'average' method assigns the rank normally first, and then calculates the average rank for the conflicting values.
# In below example, the value 4 repeats 3 times. It has ideal ranks 2,3 and 4 respectively. The 'average' method takes -
# - average of 2,3,4 which is 3, hence it assigns rank 3 to all instances of value 4.

print(s.rank(method='average'),end='\n\n')



# The 'min' method assigns the minimum rank for all conflicting values. In this example, value 4 has ideal ranks 2,3,4.
# So, it assigns rank 2 for all instances of value 4 since 2 is the minimum rank between ranks 2,3,4.
# Now, you can guess what 'max' method would do.

print(s.rank(method='min'),end='\n\n')


# The 'first' method assigns separate ranks to conflicting values. i.e. in this case, value 4 will have - 
# ranks 2,3,4 unchanged.
# It's first come first rank for conflicting values.

print(s.rank(method='first'))
```

### .apply() function

- By default, .apply() function acts on all elements element wise (sort of like map function).
- But, you can aslo perform row specific operatiins by specifying 'axis' parameter as 1. For example, if you wanted to take sum or product of all elements in a row or column.
- In case of rows, a series consisting of only 1 row is passed for axis=1, so that you can index into the series and perform operations on the elements.
- For axis = 0, an entire column is passed in the form of series, where the column name becomes the index for the row and the original indices become column names i.e. rows and columns are swapped. 

```python
# apply(series + dataframe)
temp = pd.Series([10,20,30,40,50])

temp
```

```python
def sigmoid(value):
  return 1/1+np.exp(-value)
```

```python
temp.apply(sigmoid)
```

```python
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)],
        'trial':[i for i in range(1,6)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
# Here, for axis=1, an entire row as series is passed, hence you can see label indexing in 'euclidean' functionn.

points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```

```python
# In case of axis=0, entire column is passed one at a time as a transposed series.


# Here, the variable 'srs' is a series consisting of the original column names as indices.
# and the values are the results of the function applied on values in the original column

srs = points_df[['trial','distance']].apply(sum ,axis=0)
print(srs, end='\n\n\n')


# The .to_frame() function converts the series to a df and the .T function takes the df's transpose.
# i.e. swaps the rows and columns of the df

print(srs.to_frame().T, end='\n\n\n')
```
